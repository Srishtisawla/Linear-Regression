```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression as lr
```

```python
path = 'age_religion.csv'
data = pd.read_csv(path,delimiter = ',')

#religious observance
data['religious_observance'] = data['Hours Spent_a_Week_at_Church']+data['Prayers_a_Day']+data['Meditation']+data['Hours_Spent_with_Other_Spiritual_Affairs']

#descriptive stats
age_descriptive_stats = data['Age'].describe()
religious_observance_descriptive_stats = data['religious_observance'].describe()
#print age_descriptive_stats
#print religious_observance_descriptive_stats

#graphical representation of data

#age frequency bar plot
age = data['Age'].value_counts(sort=False)
ax1 = plt.subplot(221)
age.plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Frequency barplot')
plt.show()
#age relative frequency histogram
age_total = data['Age'].sum()
relative_age = age/age_total
plt.subplot(222)
relative_age.plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Relative Frequency')
plt.title('Age Relative Frequency barplot')
plt.show()
#Religious observance frequency bar plot
religious_observance = data['religious_observance'].value_counts(sort=False)
ax1 = plt.subplot(223)
religious_observance.plot(kind='bar')
plt.xlabel('Hours of Religious_observance')
plt.ylabel('Frequency')
plt.title('religious_observance Frequency barplot')
plt.show()
#Religious observance relative frequency histogram
religious_observance_total = data['religious_observance'].sum()
relative_religious_observance = religious_observance/religious_observance_total
plt.subplot(224)
relative_religious_observance.plot(kind='bar')
plt.xlabel('Hours of Religious_observance')
plt.ylabel('Relative Frequency')
plt.title('religious_observance Relative Frequency barplot')
plt.show()
```

```python
#Linear Regression
model = lr()
model.fit(data.Age[:,np.newaxis],data['religious_observance'])
y = model.predict(data.Age[:,np.newaxis])
data['predicted_observance'] = y

plt.scatter(data['Age'],data['religious_observance'] , color='black')
plt.plot(data['Age'], data['predicted_observance'], color='blue', linewidth=2)

plt.show()
```
